```python
import os
import random
import cv2
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DIR = 'C:/Users/USER/Downloads/mid_/CIFAR-10-images-master/CIFAR-10-images-master/train'
CATEGORIES = []
for c in os.listdir(TRAIN_DIR):
    CATEGORIES.append(c)
print(CATEGORIES)
```

```python
TRAIN_DATA = []
for c in CATEGORIES:
    path = os.path.join(TRAIN_DIR,c)
    class_num = CATEGORIES.index(c)
    for img in tqdm(os.listdir(path)):
        img_arr = cv2.imread(os.path.join(path,img))
        TRAIN_DATA.append([img_arr, class_num])
print(len(TRAIN_DATA))
```

```python
random.shuffle(TRAIN_DATA)
plt.figure(figsize=(20,10))

for i in range(50):
    plt.subplot(5,10,i+1)
    plt.imshow(TRAIN_DATA[i][0])
    plt.xlabel(CATEGORIES[TRAIN_DATA[i][1]])
    plt.xticks([])
    plt.yticks([])
    if i==50:
        break
plt.show()
```

```python
f0 = TRAIN_DATA[0:1000]
f1 = TRAIN_DATA[1000:2000]
f2 = TRAIN_DATA[2000:3000]
f3 = TRAIN_DATA[3000:4000]
f4 = TRAIN_DATA[4000:5000]
```

```python


MidPoint_AccL1=[]
MidPoint_AccL2=[]


K_VALUES=[1,3,5,7,9]

for i in range(5):
    T_accL1 = []
    T_accL2 = []
    
    if i==0:
        train = f1+f2+f3+f4
        validation = f0 
    elif i==1:
        train = f0+f2+f3+f4
        validation = f1
    elif i==2:
        train = f1+f0+f3+f4
        validation = f2
    elif i==3:
        train = f1+f2+f0+f4
        validation = f3
    elif i==4:
        train = f1+f2+f3+f0
        validation = f4
        

    for k in K_VALUES:
        Acc_L1=0
        Acc_L2=0
        for v in tqdm(validation):
            List_L1 = []
            List_L2 = []
            
            
            for t in train:
                
                l2_dist = np.sum(np.sqrt(t[0]**2-v[0]**2))
                List_L2.append([l2_dist,t[1]])
                

                l1_dist = np.sum(t[0]-v[0])
                List_L1.append([l1_dist,t[1]])
            sorted_L1 = sorted(List_L1, key=lambda a: a[0])
            sorted_L2= sorted(List_L2, key=lambda a: a[0])
            

            sorted_range1 = [sorted_L1[i][1] for i in range(k)]
            sorted_range2 = [sorted_L2[i][1] for i in range(k)]
            prediction_L1 = max(set(sorted_range1), key=sorted_range1.count)
            prediction_L2 = max(set(sorted_range2), key=sorted_range2.count)

            real_label = v[1]
            if prediction_L1 == real_label:
                Acc_L1 += 1
            if prediction_L2 == real_label:
                Acc_L2 += 1



        Accuracy1 = Acc_L1 / len(validation) * 100
        Accuracy2 = Acc_L2 / len(validation) * 100
        print(f"K {k} , L1 Accuracy: {Accuracy1:.2f}%, L2 Accuracy: {Accuracy2:.2f}%")
        T_accL1.append(Accuracy1)
        T_accL2.append(Accuracy2)
        
        

    MidPoint_L2 = sum(T_accL2)/5
    MidPoint_L1 = sum(T_accL1)/5
    MidPoint_AccL1.append(MidPoint_L1)
    MidPoint_AccL2.append(MidPoint_L2)
    

fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2)

ax1.plot(K_VALUES,MidPoint_AccL1,label='For L1_accuracy')
ax1.set_xlabel('K_VALUES')
ax1.set_ylabel('Mid Accuracy')
ax1.legend(loc='best')

ax2.plot(K_VALUES,MidPoint_AccL2,label='For L2_accuracy')
ax2.set_xlabel('K_VALUES')
ax2.set_ylabel('Mid Accuracy')
ax2.legend(loc='best')


plt.tight_layout()
plt.show()
```
